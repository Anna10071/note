# 基本使用

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
from matplotlib import font_manager
```

# Matplotlib结构的三层结构（由里到外）：
第一层：Canvas（画板）：位于最底层
第二层：Figure(画布）：一个画布可以有多个axes绘图区，可以设置画布的大小和分辨率
第三层：Axes（绘图区）：数据的绘图区域
      axis（坐标轴）、legend（图例）等辅助显示层以及图像层都是建立在Axes之上

# 折线图plot的基本画法

```python
# 1、准备数据
x=range(2,26,2)
y_1=[15,13,14,17,20,25,26,26,24,12,13,11]
y_2=[11,13,12,24,26,13,15,14,17,15,12,17]

# 2、创建画布（设置画布大小和分辨率，（20，8）是指宽度和高度）
plt.figure(figsize=(20,8),dpi=80)

#3、绘制图像
plt.plot(x,y_1,color="r",linestyle="--",label="beijing")
plt.plot(x,y_2,color="b",label="shanghai")      

#显示图例和位置
plt.legend(loc=2)

#修改X、Y刻度
#如果想要步长为0.5，那么可以调用numpy库，如plt.xticks(np.arange(2,26,0.5))
plt.xticks(range(2,26,2))

#修改X的刻度说明？？？
#x_label=["11点{ }分".format(i) for i in x]
#plt.xticks(x[::2],x_label[::2])


#添加描述信息以及字体大小
plt.xlabel("Time",fontsize=18)
plt.ylabel("Wendu",fontsize=18)
plt.title("Change",fontsize=18)

#设置网格以及网格颜色的透明度
plt.grid(True,linestyle="--",alpha=0.4)

#4、保存图像
plt.savefig("text1.png")

#5、展现出图形
plt.show()
```

# 多个坐标系显示-plt.subplots(面向对象的画图方法）

```python
# 1、准备数据
x=range(2,26,2)
y_1=[15,13,14,17,20,25,26,26,24,12,13,11]
y_2=[11,13,12,24,26,13,15,14,17,15,12,17]

# 2、创建画布（设置画布大小和分辨率，（20，8）是指宽度和高度）
#修改前：plt.figure(figsize=(20,8),dpi=80)
figure,axes=plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=80)

#3、绘制图像
axes[0].plot(x,y_1,color="r",linestyle="--",label="beijing")
axes[1].plot(x,y_2,color="b",label="shanghai")      

#显示图例和位置
axes[0].legend(loc=2)
axes[1].legend(loc=2)

#修改X、Y刻度
axes[0].set_xticklabels(range(2,26,2))
axes[1].set_xticklabels(range(2,26,2))

#添加描述信息以及字体大小
axes[0].set_xlabel("Time",fontsize=18)
axes[0].set_ylabel("Wendu",fontsize=18)
axes[0].set_title("Change",fontsize=18)

axes[1].set_xlabel("Time",fontsize=18)
axes[1].set_ylabel("Wendu",fontsize=18)
axes[1].set_title("Change",fontsize=18)

#设置网格以及网格颜色的透明度
axes[0].grid(True,linestyle="--",alpha=0.4)
axes[1].grid(True,linestyle="--",alpha=0.4)

#4、展现出图形
plt.show()
```

## 绘制数学函数图像

```python
#1、准备数据x、y
x=np.linspace(-10,10,1000)
y=2*x*x

#2、创建画布
plt.figure(figsize=(20,8),dpi=80)

#3、绘制图像
plt.plot(x,y)

#4、绘制图像
plt.show()
```

# 常见的图形以及意义：
1、折线图plot：
2、散点图scatter：探究不同变量之间的关系以及规律；
3、柱状图bar：能够一眼看出各个数据的大小，比较数据之间的差别；
4、直方图histogram：展示数据的分布；观察和估计哪些数据比较集中，异常或者孤立的数据分布在何处；
